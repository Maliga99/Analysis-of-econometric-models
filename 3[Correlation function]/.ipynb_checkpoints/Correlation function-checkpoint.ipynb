# Задание №1
### *Построение автокорреляционной функции*

```python
import numpy as np

data = np.array([7.5, 7.2, 6.1, 6.8, 7.4, 7.8, 7.5, 7.1, 7.1, 7.4, 7.0, 6.4, 7.4, 6.8, 6.7, 6.0, 6.3, 6.2, 6.7, 5.6, 6.1, 5.1, 5.6, 6.6, 6.7, 5.9, 6.9, 5.4, 6.2, 5.3, 5.4, 6.7, 6.2, 6.6, 6.3, 5.8, 6.8])
```

### 1.1.
##### Вычисление автокорреляционной функции $r_s$

```python
def autocorr_function(T, s, data):
    return (np.dot(data[0:T-s], data[s:]) - sum(data[0:T-s] * sum(data[s:] / (T-s)))) / np.sqrt((sum(data[0:T-s] ** 2) - sum(data[0:T-s]) ** 2 / (T-s))) / np.sqrt((sum(data[s:] ** 2) - sum(data[s:]) ** 2 / (T-s)))


T = data.size
s_max = 20
S = np.arange(0, s_max+1)

r_s = []
for s in S:
    r_s.append(autocorr_function(T, s, data))
```

### 1.2.
##### Построение коррелограммы

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.plot(S, r_s,'r-*')
plt.xlabel('$S$', fontsize=15)
plt.ylabel('$r_s$', fontsize=15)
plt.title('Коррелограмма', fontsize=15)
plt.xticks(np.arange(0, s_max+1, 2))
plt.grid(ls='--')
plt.show()
```

### *Анализ временной изменчивости ряда*

### 2.1.
##### Метод серий, метод восходящих и нисходящих серий для проверки наличия тренда

```python
import math

def search(sign):
    p = [1]
    for i, el in enumerate(sign[1:]):
        if el == sign[i]:
            p[-1] += 1
        else:
            p.append(1)
    
    return p

            
median = np.median(data)
sign = np.delete(data, np.where(data == median)) > median


seq = search(sign)
v, t = len(seq), max(seq)

print('Проверяем неравенства:', v > math.floor(0.5 * (data.size + 1 - 1.96 * np.sqrt(data.size - 1))), t < math.floor(3.3 * (math.log10(data.size) + 1))) 
```

Таким образом по **методу серий** гипотеза об отсутствии тренда отвергается.

```python
sign = np.zeros(data.size)

if data[-1] > 0:
    sign[-1] = 1

for i in np.arange(0, data.size-1):
    if data[i+1] - data[i] > 0:
        sign[i] = 1
    elif data[i+1] - data[i] == 0 and data[i] > 0:
        sign[i] = 1

        
seq = search(sign)
v, t = len(seq), max(seq)

print('Проверяем неравенства:', v > math.floor(1/3 * (2 * data.size - 1) - 1.96 * np.sqrt((16 * data.size - 29) / 90)), t < 6)
```

Таким образом по **методу восходящих и нисходящих серий** гипотеза о случайности выборки не может быть отвергнута.

### 2.2.
##### Сглаживание ряда динамики методом скользящего среднего 

```python
l = 7
dynamic_range = np.zeros(data.size - l + 1)

for i, j in enumerate(np.arange(l, data.size+1)):
    dynamic_range[i] = data[i:j].mean()

    
plt.figure(figsize=(12,8))

plt.plot(np.arange(0, data.size), data,'r-*', label='Range')
plt.plot(np.arange(3, dynamic_range.size + 3), dynamic_range,'b-.o', label='MA')
plt.xlabel('$t$', fontsize=15)
plt.ylabel('$T$', fontsize=15)
plt.title('График ряда и сглаженного ряда', fontsize=15)
plt.xticks(np.arange(0, data.size, 2))
plt.legend(fontsize=15)
plt.grid(ls='-')
plt.show()
```

### 2.3.
##### Рассчет линейной модели тренда

```python
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()
f = linear_regressor.fit(np.arange(0, data.size).reshape(-1, 1), data)

print("y = {:.2f}*x + {:.2f}".format(linear_regressor.coef_[0], linear_regressor.intercept_))
```

### 2.4.
##### Рассчет коэфф-та корреляции $r_{ty}$, его стандартной ошибки $\sigma_r$ и коэфф-та детерминации $\eta_{y(t)}$

```python
t = np.arange(0, data.size)

r_ty = np.corrcoef(t, data)[0][1]
sigma_r = np.sqrt((1 - r_ty ** 2) / (data.size - 2))
nu_2 = 1 / data.size * sum((f.predict(t.reshape(-1, 1)) -  data.mean()) ** 2) / data.var() 
print('Коэф. корреляции : {}'.format(r_ty))
print('Cтандартная ошибка : {}'.format(sigma_r))
print('Коэф. детерминации : {}'.format(nu_2))
```

### 2.5.
#####  Оценка значимости коэф-та корреляции через критерий Стьюдента

```python
from scipy.stats import t

t_stud = r_ty / sigma_r

alpha = 0.025

print('t_кр для alpha = {} : {}'.format(alpha, t(data.size).ppf(1-alpha)))
print('t_рас :', abs(t_stud))
print(abs(t_stud) < t(data.size).ppf(1-alpha))
```

### 2.6.
##### Вычисление дисперсии отклонения $\sigma^2_{\epsilon}$ 

```python
sigma_eps = data.var() * (1 - r_ty ** 2)
print('Дисперсия отклонения: {}'.format(sigma_eps))
```
