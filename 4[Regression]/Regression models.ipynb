### Задание 2
##### Разделение данных для тренировки (80%), валидации (10%) и тестирования (10%) модели

Импортируем необходимую библиотеку для разделения данных

```python
from sklearn.model_selection import train_test_split
```

Разобьем данные на тест и тренировку

```python
X_train_v, X_test, y_train_v, y_test = train_test_split(df.iloc[:,:-1].values, df.iloc[:,-1].values, test_size=0.1)
```

Разобьем `train_v` на тренировочные и валидационные данные (0.111 здесь потому что мы берем 10% не от 100%, а от 90%)

```python
X_train, X_val, y_train, y_val = train_test_split(X_train_v, y_train_v, test_size=0.111)
```

### Задание 3
##### Отбор признаков для модели линейной регрессии:  `forward` и `backward` feature selection.

Импортируем необходимые модули

```python
from sklearn.feature_selection import RFE
from mlxtend.feature_selection import SequentialFeatureSelector
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
```

Построим модель линейной регрессии

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lr1 = LinearRegression()
lr2 = LinearRegression()
```

Отберем 15 признаков

```python
N = 15
```

#### Forward feature selection

```python
sfs = SequentialFeatureSelector(lr1, k_features=N, forward=True, scoring='neg_mean_squared_error', cv=0)

sfs.fit(X_train, y_train)
```

Теперь обучим линейную регрессию на отобранных 50 признаках

```python
lr1.fit(X_train[:, np.array(sfs.k_feature_idx_)], y_train)

print('Среднеквадратичная ошибка: ', mean_squared_error(y_val, lr1.predict(X_val[:, np.array(sfs.k_feature_idx_)])))
print('Коэффициент детерминации: ', r2_score(y_val, lr1.predict(X_val[:, np.array(sfs.k_feature_idx_)])))
```

##### Backward feature selection

```python
rfe = RFE(lr2, n_features_to_select=N)

rfe.fit(X_train, y_train)
```

Теперь обучим линейную регрессию на отобранных 50 признаках

```python
lr2.fit(X_train[:, rfe.get_support()], y_train)

print('Среднеквадратичная ошибка: ', mean_squared_error(y_val, lr2.predict(X_val[:, rfe.get_support()])))
print('Коэффициент детерминации: ', r2_score(y_val, lr2.predict(X_val[:, rfe.get_support()])))
```

Данные очень плохо описываются линейными моделями, но первая модель немного лучше

### Задание 4 
##### Кросс-валидация на K фолдов

```python
from sklearn.model_selection import cross_val_score

k3 = cross_val_score(lr1, X_train[:, np.array(sfs.k_feature_idx_)], y_train, cv=3).mean()
k5 = cross_val_score(lr1, X_train[:, np.array(sfs.k_feature_idx_)], y_train, cv=5).mean()
k10 = cross_val_score(lr1, X_train[:, np.array(sfs.k_feature_idx_)], y_train, cv=10).mean()

k3, k5, k10
```

Кросс-валидация на 10 фолдах дает лучшую среднюю оценку на кросс-валидации 

### Задание 5
##### Построим график c помощью наилучшей модели

```python
plt.figure(figsize=(20, 8))

plt.plot(np.arange(0, *y_test.shape), y_test, ':', label='Фактические')
plt.plot(np.arange(0, *y_test.shape), lr1.predict(X_test[:, np.array(sfs.k_feature_idx_)]), label='Прогнозируемые')
plt.title('Графики фактических и прогнозируемых значений', fontsize=20)
plt.ylabel('$y$', fontsize=20)
plt.legend(fontsize=20)
plt.show()
```

### Задание 7
##### RidgeRegression

$||y - Xw||^2_2 + \alpha * ||w||^2_2$

```python
from sklearn.linear_model import Ridge
```

Параметры $\alpha$ для регуляризаторов

```python
alpha_1 = 0.001
alpha_2 = 0.01
alpha_3 = 0.1
```

Теперь создадим 3 модели с разными параметрами $\alpha$

```python
ridge_reg_1 = Ridge(alpha=alpha_1)
ridge_reg_2 = Ridge(alpha=alpha_2)
ridge_reg_3 = Ridge(alpha=alpha_3)
```

```python
ridge_reg_1.fit(X_train, y_train)
ridge_reg_2.fit(X_train, y_train)
ridge_reg_3.fit(X_train, y_train)
```

### Задание 9
