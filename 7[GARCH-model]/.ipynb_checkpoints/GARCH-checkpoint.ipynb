### Задание №2
##### GARCH(1,1) для доходности ряда

Вычислим логарифмическую доходность для ряда.

$r_t = log(\frac{S_t}{S_{t-1}}) $ - логарифмическая доходность.

```python
import numpy as np

data = data.values.reshape(-1)
```

```python
r_t = np.diff(np.log(data))
r_t
```

Оценим нормальную GARCH(1,1)-модель.

```python
from arch import arch_model

model = arch_model(r_t, mean='Zero', vol='GARCH', p=1, q=1, dist='normal')
res = model.fit()
res.summary()
```

Исследуем полученную лог. доходность на стационарность. Для проверки стационарности проведем обобщенный тест Дикки-Фуллера на наличие единичных корней. 

```python
import statsmodels as sm

test_DF = sm.tsa.stattools.adfuller(r_t)

print('adf: ', test_DF[0])
print('p-value: ', test_DF[1])
print('Critical values: ', test_DF[4])
if test_DF[0] > test_DF[4]['5%']: 
    print('Есть единичные корни, ряд не стационарен')
else:
    print('Единичных корней нет, ряд стационарен')
```

Вычислим автокорреляционную функцию квадратов GARCH(1,1).

```python
auto_corr_sigma = sm.tsa.stattools.acf(res.conditional_volatility ** 2, nlags=n-1)
```

Построим график автокорреляции квадратов GARCH(1,1).

```python
plot_acf(res.conditional_volatility ** 2)
plt.show()
```

### Задание №3
##### Построить график условной дисперсии. Вычислить нормированные остатки, исследовать их, построить автокорреляционную функцию. 

Построим график условной дисперсии.

```python
plt.figure(figsize=(15, 6))

plt.plot(res.conditional_volatility ** 2, label='$\sigma_t^2$')
plt.xlabel('$t$', fontsize=20)
plt.ylabel('Variance', fontsize=20)
plt.title('Сonditional variance', fontsize=25)
plt.legend(fontsize=15)
plt.show()
```

Вычислим нормированные остатки.

$h_t = \sigma_t\varepsilon_t \Rightarrow \varepsilon_t = \frac{h_t}{\sigma_t}$

```python
normalize_resid = r_t / res.conditional_volatility
```

Исследуем нормированные остатки на независимость с помощью h-теста Дарбина.

$h = (1 - \frac{1}{2}DW) \sqrt{\frac{n}{1 - nV}}$, где $V$ - это квадрат стандартной ошибки при лаговой факторной переменной $y_{t-1}$, $DW$ - статистика Дарбина-Уотсона.

```python
import statsmodels.api as sm

h = (1 - sm.stats.stattools.durbin_watson(normalize_resid) / 2) * np.sqrt(normalize_resid.shape[0] / (1 - normalize_resid.shape[0] * res.std_err['alpha[1]'] ** 2))
h
```

При увеличении объёма выборки распределение h-статистики стремится к нормальному с нулевым математическим ожиданием и дисперсией, равной 1. Поэтому гипотеза об отсутствии автокорреляции остатков отвергается, т.к. фактическое значение 
h-статистики оказалось больше, чем критическое значение нормального распределения $\alpha=1.96$.

Исследуем нормированные остатки на соответствие нормальному распределению с помощью  теста Шапиро-Уиллока.

```python
from scipy import stats

sw_test = stats.shapiro(normalize_resid)
sw_test
```

$p-value < \alpha=0.05$, следовательно мы можем отвергнуть гипотезу о нормальности распределения. 

Построим автокорреляционную функцию нормированных остатков.

```python
plot_acf(normalize_resid)
plt.show()
```

### Задание  №4
##### Построить условный доверительный интервал прогнозов GARCH(1,1)-процесса. Изобразить графически.

Рассчитаем дисперсии ошибок прогноза.

$\begin{cases}
   \sigma^2_{d_1} = \sigma^2_{2}
   \\
   \sigma^2_{d_k} = \omega + (\alpha_1 + \beta_1)\sigma^2_{d_{k-1}}.
 \end{cases}$

После преобразования получим:

$\sigma^2_{d_k} = \omega \frac{1 - (\alpha_1 + \beta_1)^{k-1}}{1 - \alpha_1 - \beta_1} +  (\alpha_1 + \beta_1)^{k-1} \sigma^2_{2}$.

```python
def forecast_error(sigma, omega, alpha, beta, k):
    return omega * (1 - (alpha + beta) ** (k-1)) / (1 - alpha - beta) + (alpha + beta) ** (k-1) * sigma


omega, alpha, beta = res.params.values
sigma = res.conditional_volatility[1]

error_std = np.sqrt(np.array([forecast_error(sigma, omega, alpha, beta, k) for k in range(1, res.conditional_volatility.shape[0] + 1)]))
```

Построим верхнюю и нижнюю границы интервала.

```python
up_border = res.conditional_volatility + 2 * error_std

down_border = -res.conditional_volatility - 2 * error_std
```

Изобразим истинные значения и интервальный прогноз.

```python
plt.figure(figsize=(15, 6))

plt.plot(up_border)
plt.plot(down_border)
plt.plot(r_t, label='Data')
plt.xlabel('$t$', fontsize=20)
plt.ylabel('$h$', fontsize=20)
plt.title('Prediction', fontsize=25)
plt.legend(fontsize=20)
plt.show()
```
